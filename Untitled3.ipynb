```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm
import random

d = datasets.load_iris()          # iris 데이터셋 불러오기

# iris 데이터셋을 훈련 집합 80%, 테스트 집합 20%로 나누기
X_train, X_test, y_train, y_test = train_test_split(d.data, d.target, test_size = 0.2, random_state = 0)

s = svm.SVC(gamma=0.01, C=10).fit(X_train, y_train)   # SVM 모델 학습하기

res = s.predict(X_test)           # 예측하기
score = s.score(X_test, y_test)   # SVM 성능 확인하기

print(res)     # 예측한 부류 출력
print(score)   # SVM 성능 출력





```
